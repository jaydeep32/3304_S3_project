```python
# Authenticate with Twitter API
auth = tweepy.OAuth1UserHandler(api_key, api_secret_key, access_token, access_token_secret)
api = tweepy.API(auth)
```

```python

def fetch_tweets_with_location(query, count=100):
    tweets_data = []
    for tweet in tweepy.Cursor(api.search_tweets, q=query, count=count, tweet_mode='extended').items(count):
        if tweet.coordinates:
            tweets_data.append({
                'text': tweet.full_text,
                'lat': tweet.coordinates['coordinates'][1],
                'lon': tweet.coordinates['coordinates'][0],
                'created_at': tweet.created_at
            })
    return pd.DataFrame(tweets_data)
```

```python
# Fetch tweets with location for a specific query
query = "Tech"
tweets_df = fetch_tweets_with_location(query, count=100)

# Plot tweets on a map using Folium
m = folium.Map(location=[0, 0], zoom_start=2)
marker_cluster = MarkerCluster().add_to(m)

for idx, row in tweets_df.iterrows():
    folium.Marker(location=[row['lat'], row['lon']], popup=row['text']).add_to(marker_cluster)

display(m)
```
